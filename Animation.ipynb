```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML
from matplotlib import style

fig, ax=plt.subplots()
ax.set_xlim((0,1))
ax.set_ylim((-1.1))
ax.set_xlabel("$x$", fontsize=12)
ax.set_ylabel("$y$", fontsize=12)
ax.set_title("sin animation")
ax.grid(True)

line, =ax.plot([],[],lw=2)

def init():
  line.set_data([],[])
  return (line,)
  
def animate(t):
  x=np.linspace(0,1,1000)
  y=np.sin(2*np.pi*x)*np.exp(-0.1*t)
  line.set_data(x,y)
  return (line,)

ani = animation.FuncAnimation(fig, animate, init_func=init, frames=100, interval=120, blit=True)
HTML(ani.to_html5_video())
```
